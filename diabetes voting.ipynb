```python
x= df.drop('Outcome', axis=1)
```

```python
y= df['Outcome']
```

```python
scale= StandardScaler()
```

```python
x= scale.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_test.shape
```

# machine learning

```python
# logistic regression
log= LogisticRegression()
```

```python
log.fit(x_train, y_train)
```

```python
logpred= log.predict(x_test)
```

```python
logpred [:10]
```

```python
logacc= accuracy_score(y_test, logpred)*100
```

```python
logacc
```

```python
logclass= classification_report(y_test, logpred)
```

```python
dtc= DecisionTreeClassifier()
```

```python
dtc.fit(x_train, y_train)
```

```python
dtcpred= dtc.predict(x_test)
```

```python
dtcpred [:10]
```

```python
dtcacc= accuracy_score(y_test, dtcpred)*100
```

```python
dtcacc
```

```python
dtclass= classification_report(y_test,dtcpred)
```

```python
print(dtclass)
```

```python
rfc= RandomForestClassifier()
```

```python
rfc.fit(x_train, y_train)
```

```python
rfcpred= rfc.predict(x_test)
```

```python
rfcpred [:10]
```

```python
rfclass= classification_report(y_test, rfcpred)
```

```python
print(rfclass)
```

```python
svm= SVC(probability=True)
```

```python
svm.fit(x_train, y_train)
```

```python
svmpred= svm.predict(x_test)
```

```python
svmpred [:10]
```

```python
svmacc= accuracy_score(y_test, svmpred)*100
```

```python
svmacc
```

```python
svmclass= classification_report(y_test, svmpred)
```

```python
print(svmclass)
```

```python
neighbors= KNeighborsClassifier()
```

```python
neighbors.fit(x_train, y_train)
```

```python
neighborspred= neighbors.predict(x_test)
```

```python
neighborspred [:10]
```

```python
neighborsacc= accuracy_score(y_test, neighborspred)*100
```

```python
neighborsacc
```

```python
voting= VotingClassifier(estimators=[('log', log), ('dtc', dtc), ('rfc', rfc), ('svm', svm), ('neighbor', neighbors)], voting='soft')
```

```python
voting.fit(x_train, y_train)
```

```python
votingpred= voting.predict(x_test)
```

```python
votingpred [:10]
```

```python
votingacc= accuracy_score(y_test, votingpred)*100
```

```python
votingacc
```
